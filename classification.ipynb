```python
import pandas as pd
import numpy as np
from tqdm import tqdm
from transformers import pipeline
```

```python
category_mapping = {
    "Fiction": "Fiction",
    "Juvenile Fiction": "Children Fiction",
    "Biography & Autobiography": "Nonfiction",
    "History": "Nonfiction",
    "Philosophy": "Nonfiction",
    "Religion": "Nonfiction",
    "Science": "Nonfiction",
    "Poetry": "Nonfiction",
    "Drama": "Fiction",
    "Comics & Graphic Novels": "Fiction",
    "Juvenile Nonfiction": "Children Nonfiction",
    "Literary Criticism": "Nonfiction",
}
```

```python
df['new_categories'] = df['categories'].map(category_mapping)
df
```

```python
df[~(df['new_categories'].isna())]
```

```python
pipe = pipeline('zero-shot-classification', model='facebook/bart-large-mnli')
```

```python
pipe.save_pretrained('./bart-large-mnli')
```

```python
fiction_categories = ['Fiction','Nonfiction']
sequence = df.loc[df['new_categories'] == 'Fiction', 'description'].reset_index(drop=True)[0]
sequence
```

```python
pipe(sequence, fiction_categories)
```

```python
max_idx = np.argmax(pipe(sequence, fiction_categories)['scores'])
label = pipe(sequence, fiction_categories)['labels'][max_idx]
label
```

```python
def predict_category(sequence,categories):
    pred = pipe(sequence, categories)
    max_idx = np.argmax(pipe(sequence, categories)['scores'])
    label = pred['labels'][max_idx]
    return label
```

```python
actual_label = []
predicted_label = []

for i in tqdm(range(0,300)):
    sequence = df.loc[df['new_categories'] == 'Fiction', 'description'].reset_index(drop=True)[i]
    actual_label.append('Fiction')
    predicted_label.append(predict_category(sequence))
```

```python
for i in tqdm(range(0,300)):
    sequence = df.loc[df['new_categories'] == 'Nonfiction', 'description'].reset_index(drop=True)[i]
    actual_label.append('Nonfiction')
    predicted_label.append(predict_category(sequence))
```

```python
predictions = pd.DataFrame({'actual_label': actual_label, 'predicted_label': predicted_label})
predictions
```

```python
correct_predictions = predictions[predictions['actual_label'] == predictions['predicted_label']].shape[0]
correct_predictions
```

```python
accuracy = correct_predictions / predictions.shape[0]
accuracy
```

```python
isbn = []
predicted_category = []

missing_categories = df.loc[df['new_categories'].isna(), ['isbn13','description']].reset_index(drop=True)

for i in tqdm(range(0,len(missing_categories))):
    isbn.append(missing_categories['isbn13'][i])
    predicted_category.append(predict_category(missing_categories['description'][i], fiction_categories))
```

```python
predictions = pd.DataFrame({'isbn13': isbn, 'predicted_category': predicted_category})
predictions
```

```python
predictions.to_csv('missing_categories_predictions.csv', index=False)
```
